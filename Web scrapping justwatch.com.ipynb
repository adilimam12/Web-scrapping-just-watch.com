# Most Popular Movie scrapping

```python
browser = webdriver.Chrome()
```

```python
url="https://www.justwatch.com/in/movies?genres=act,rma,scf&release_year_from=2020"
```

```python
#open the page url in chrome
browser.get(url)
```

```python
#Scrapping the title
title_list = browser.find_elements(By.XPATH,"//a[@class='title-list-grid__item--link']")
title_list

```

```python
#Storing the movie url in a list
movies=[]
for i in title_list:
    name=i.get_attribute('href')
    movies.append(name)
movies
```

```python
# Function to extract movie name from the URL
def extract_movie_name_from_url(url):
    return url.split('/')[-1]

# Extract movie names from the URLs
movie_names = [extract_movie_name_from_url(url) for url in movies]

# Print the extracted movie names
print(movie_names)
```

```python
Release_year=[]
for movie_url in movies:
    # Navigate to the movie URL
    browser.get(movie_url)

    try:
        # Find the release year element
        release_year_element = browser.find_element(By.XPATH, "//span[@class='text-muted']")
        release_year = release_year_element.text.strip("()")
    except:
        release_year = "Release Year not found"
    Release_year.append({"Movie URL": movie_url, "Release Year": release_year})
    print(Release_year)
```

```python
# Extract release years from the Release_year list
movie_release_years = [year["Release Year"] for year in Release_year]

# Display release years
print(movie_release_years)
```

```python
#Scrapping the movie imdb ratings
movie_data = []

for movie_url in movies:
    response = requests.get(movie_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        details_div = soup.find('div', {'class': 'detail-infos'})

        # Scrape the required data from 'details_div'
        imdb_rating = details_div.find('div', {'class': 'detail-infos__value'}).text.strip()
      

        # Append the scraped data to the 'movie_data' list
        movie_data.append({
            'url': movie_url,
            
            'imdb_rating': imdb_rating,
            
        })

# Print the scraped data for all movies
for data in movie_data:
    print(f"Movie URL: {data['url']}")
    
    print(f"IMDb Rating: {data['imdb_rating']}")
    
    print("\n")
```

```python
movie_data
```

```python
#imdb ratings movie
movie_imdb_ratings = []
for movie in movie_data:
    imdb_rating = movie["imdb_rating"]
    imdb_rating = imdb_rating.replace("%", "").strip().split()[1]
    movie_imdb_ratings.append(imdb_rating)

print(movie_imdb_ratings)
```

```python
#scrapping genre
def fetch_genre(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the genres using the provided XPath expression
        genres_heading = soup.find('h3', class_='detail-infos__subheading', text='Genres')
        if genres_heading:
            genres_element = genres_heading.find_next_sibling('div', class_='detail-infos__value')
            genres = genres_element.text.strip()

            # Split the genres into a list
            genres_list = [genre.strip() for genre in genres.split(',')]
            return genres_list
    return None

# Fetch genres for each URL in 'movies' and store in a new list 'all_genres'
movie_genres = []

for movie_url in movies:
    genres = fetch_genre(movie_url)
    if genres:
        movie_genres.extend(genres)

# Print the list of all genres
print(movie_genres)
```

```python
#Scrapping straming service provider 
Provider = []  # Use the same list variable consistently

for movie_url in movies:
    # Navigate to the movie URL
    browser.get(movie_url)

    try:
        # Find the provider element
        provider_element = browser.find_element(By.XPATH, "//img[@class='offer__icon']")
        provider_name = provider_element.get_attribute('alt')
    except:
        provider_name = "Provider not found"

    Provider.append({"Movie URL": movie_url, "Provider Name": provider_name})
    print(Provider)  # Print the Provider list
```

```python
Movie_streaming = [movie_info['Provider Name'] for movie_info in Provider]

print(Movie_streaming)
```

# Tv Show scrapping

```python
#defining a new variable for scrapping tv show data
url2="https://www.justwatch.com/in/tv-shows?release_year_from=2020" 
```

```python
browser.get(url2)
```

```python
title_tv_list=browser.find_elements(By.XPATH,"//a[@class='title-list-grid__item--link']")
title_tv_list
```

```python
#Storing the tv show url in a list
Tv_url=[]
for i in title_tv_list:
    name=i.get_attribute('href')
    Tv_url.append(name)
Tv_url
```

```python
#Tv show titles
tv_show_names = []

# Extract the TV show names from the URLs
for url in Tv_url:
    tv_show_name = url.split('/')[-1]
    tv_show_names.append(tv_show_name)

print(tv_show_names)
```

```python
#Scrapping release year for the show
Tv_Release_year=[]
for tv_url in Tv_url:
    # Navigate to the Tv URL
    browser.get(tv_url)

    try:
        # Find the release year element
        release_year_element = browser.find_element(By.XPATH, "//span[@class='text-muted']")
        release_year = release_year_element.text.strip("()")
    except:
        release_year = "Release Year not found"
    Tv_Release_year.append({"Tv url": tv_url, "Release Year": release_year})
    print(Tv_Release_year)
```

```python
# Extract release years from the Release_year list
tv_release_years = [year["Release Year"] for year in Tv_Release_year]

# Display release years
print(tv_release_years)
```

```python
#Scrapping the tv imdb ratings
Tv_data = []

for tv_url in Tv_url:
    response = requests.get(tv_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        details_div = soup.find('div', {'class': 'detail-infos'})

        # Scrape the required data from 'details_div'
        imdb_rating = details_div.find('div', {'class': 'detail-infos__value'}).text.strip()
      

        # Append the scraped data to the 'Tv_data' list
        Tv_data.append({
            'url': tv_url,
            
            'imdb_rating': imdb_rating,
            
        })

# Print the scraped data for all Tv shows
for data in Tv_data:
    print(f"Tv URL: {data['url']}")
    
    print(f"IMDb Rating: {data['imdb_rating']}")
    
    print("\n")
```

```python
Tv_data
```

```python
#Extracting imdb ratings for the show 
import re
Tv_imdb_ratings = []

# Define a regular expression pattern to extract IMDb ratings
imdb_pattern = r'\d+\.\d+'

# Extract the IMDb ratings from the list of dictionaries
for data in Tv_data:
    imdb_rating = data['imdb_rating']
    matches = re.findall(imdb_pattern, imdb_rating)
    if matches:
        Tv_imdb_ratings.append(matches[0])

print(Tv_imdb_ratings)
```

```python
#scrapping genre
def fetch_genre(url2):
    response = requests.get(url2)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the genres using the provided XPath expression
        genres_heading = soup.find('h3', class_='detail-infos__subheading', text='Genres')
        if genres_heading:
            genres_element = genres_heading.find_next_sibling('div', class_='detail-infos__value')
            genres = genres_element.text.strip()

            # Split the genres into a list
            genres_list = [genre.strip() for genre in genres.split(',')]
            return genres_list
    return None

# Fetch genres for each URL in 'tv' and store in a new list 'Tv_generes'
Tv_genres = []

for tv_url in Tv_url:
    genres = fetch_genre(tv_url)
    if genres:
        Tv_genres.extend(genres)

# Print the list of all genres
print(Tv_genres)
```

```python
#Scrapping straming service provider 
Streaming_service = []  # Use the same list variable consistently

for tv_url in Tv_url:
    # Navigate to the Tv URL
    browser.get(tv_url)

    try:
        # Find the provider element
        provider_element = browser.find_element(By.XPATH, "//img[@class='offer__icon']")
        provider_name = provider_element.get_attribute('alt')
    except:
        provider_name = "Provider not found"

    Streaming_service.append({"Tv URL": tv_url, "Provider Name": provider_name})
    print(Streaming_service)  # Print the Provider list
```

# Data filtering & analysis

```python
tv_data_filtered = df2[df2['Release year'] >=2023-2]
tv_data_filtered
filtered_tv_data_df=tv_data_filtered.copy()
filtered_tv_data_df.head()
```

```python
df1['Release Year'] = pd.to_numeric(df1['Release Year'], errors='coerce')
```

```python
movie_data_filtered = df1[df1['Release Year'] >=2023 - 2]
movie_data_filtered
```

```python
# Convert IMDb ratings to numeric, and replace "not found" values with NaN
df1['Movie IMDb Ratings'] = pd.to_numeric(df1['Movie IMDb Ratings'], errors='coerce')

# Calculate the average IMDb rating for movies
average_movie_imdb_rating = df1['Movie IMDb Ratings'].mean()

print("Average IMDb rating for movies in df1:", average_movie_imdb_rating)
```

```python
# Clean the IMDb Ratings column by extracting only the numeric parts and converting to float
df2['Imdb Ratings'] = df2['Imdb Ratings'].str.extract(r'(\d+\.\d+)').astype(float)

# Replace "not found" values with NaN
df2['Imdb Ratings'] = df2['Imdb Ratings'].replace('not found', float('nan'))

# Convert 'Imdb Ratings' column to numeric (float)
df2['Imdb Ratings'] = pd.to_numeric(df2['Imdb Ratings'], errors='coerce')

# Fill NaN values with 0
df2['Imdb Ratings'] = df2['Imdb Ratings'].fillna(0)

# Calculate the average IMDb rating for TV shows in df2
average_imdb_rating_tv = df2['Imdb Ratings'].mean()

print("Average IMDb rating for TV shows in df2:", average_imdb_rating_tv)
```

```python
# Assuming df1 and df2 are your DataFrames containing movie and TV show data respectively

# Convert the IMDb ratings columns to numeric values (floats)
df1['Movie IMDb Ratings'] = pd.to_numeric(df1['Movie IMDb Ratings'], errors='coerce')
df2['Imdb Ratings'] = pd.to_numeric(df2['Imdb Ratings'], errors='coerce')

# Filter movies with IMDb rating 7 or higher
movies_high_rating = df1[df1['Movie IMDb Ratings'] >= 7]


# Filter TV shows with IMDb rating 7 or higher
tv_shows_high_rating = df2[df2['Imdb Ratings'] >= 7]
movies_high_rating.head()
tv_shows_high_rating.head()
# # Save the filtered data into new DataFrames
# movies_high_rating.to_csv('movies_high_rating.csv', index=False)
# tv_shows_high_rating.to_csv('tv_shows_high_rating.csv', index=False)
```

```python
# Assuming movie_data(df1) and tv_data(df2) are the DataFrames containing movie and TV show data respectively

# Combine movie_data and tv_data into a single DataFrame with a common column name for genres
df1['Genre'] = df1['Movie Genres']
df2['Genre'] = df2['Genere']
combined_data = pd.concat([df1[['Genre']],
                          df2[['Genre']]])

# Group the data by the 'Genre' column and count the occurrences of each genre
genre_counts = combined_data['Genre'].value_counts()

# Sort the genres in descending order based on their counts
sorted_genres = genre_counts.sort_values(ascending=False)

# Select the top 5 genres
top_5_genres = sorted_genres.head(5)

print("Top 5 genres with the highest number of available movies and TV shows:")
print(top_5_genres)
```

```python
# Assuming movie_data and tv_data are the DataFrames containing movie and TV show data respectively

# Combine movie_data and tv_data into a single DataFrame with a common column name for streaming service
df1['Streaming Service'] = df1['Movie Streaming Service']
df2['Streaming Service'] =df2['streaming service']
combined_data = pd.concat([df1[['Streaming Service']],
                          df2[['Streaming Service']]])

# Group the data by the 'Streaming Service' column and count the occurrences of each streaming service
service_counts = combined_data['Streaming Service'].value_counts()

# Find the streaming service with the most number of offerings
# most_offering_service = service_counts.idxmax()
# count_of_most_offering_service = service_counts.max()
service_counts = combined_data['Streaming Service'].value_counts()

# print("Streaming service with the most number of offerings:")
# print(f"{most_offering_service} - {count_of_most_offering_service} offerings")
# Convert the series to a DataFrame
service_counts_df = service_counts.reset_index()
service_counts_df.columns = ['Streaming Service', 'Number of Offerings']

print("Streaming services with the number of offerings:")
print(service_counts_df)
```

```python
# Create DataFrames from numpy.float64 values
average_movie_imdb_rating_df = pd.DataFrame({'Average IMDb Rating Movies': [average_movie_imdb_rating]})
average_imdb_rating_tv_df = pd.DataFrame({'Average IMDb Rating TV Shows': [average_imdb_rating_tv]})

# Create an Excel writer
writer = pd.ExcelWriter('analysis_results.xlsx', engine='xlsxwriter')

# Write each DataFrame to a separate sheet in the Excel file
movie_data_filtered.to_excel(writer, sheet_name='Movie Data Filtered', index=False)
tv_data_filtered.to_excel(writer, sheet_name='TV Data Filtered', index=False)
average_movie_imdb_rating_df.to_excel(writer, sheet_name='Average IMDb Rating Movies', index=False)
average_imdb_rating_tv_df.to_excel(writer, sheet_name='Average IMDb Rating TV Shows', index=False)
movies_high_rating.to_excel(writer, sheet_name='Movies with High Rating', index=False)
tv_shows_high_rating.to_excel(writer, sheet_name='TV Shows with High Rating', index=False)
top_5_genres.to_excel(writer, sheet_name='Top 5 Genres', index=True)
service_counts_df.to_excel(writer, sheet_name='Streaming Service Counts', index=True)

# Save the Excel file
writer.save()
```
